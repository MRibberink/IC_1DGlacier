```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("MATPLOTLIB_RCPARAMS.sty")
matplotlib.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f.set_figwidth(15)
#plt.title('Response times for growing and shrinking glaciers')
plt.style.use("MATPLOTLIB_RCPARAMS.sty")
matplotlib.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
rt_up=np.load('final_rt_up.npy',allow_pickle=True).item()
rt_down=np.load('final_rt_down.npy',allow_pickle=True).item()
labellist1=[r'$1850m \rightarrow 1800m$',r'$1800m \rightarrow 1750m$',r'$1750m \rightarrow 1700m$',r'$1700m \rightarrow 1650m$',r'$1650m \rightarrow 1600m$']
labellist2=[r'$1800m \rightarrow 1850m$',r'$1750m \rightarrow 1800m$',r'$1700m \rightarrow 1750m$',r'$1650m \rightarrow 1700m$',r'$1600m \rightarrow 1650m$']
x=[float(e) for e in rt_up.keys()]
#ax1.title.set_text(r'Response times for growing glacier')
for i in range(len(rt_up['0.005'])):
    ax1.plot(x,[rt_up[e][i] for e in rt_up.keys()],label=labellist1[i])
ax1.set_xlabel(r'$\beta~[yr^{-1}]$')
ax1.set_ylabel(r'$\tau~[yr]$')

#ax1.set_title('Response times for growing glacier')
ax1.legend()
for i in range(len(rt_up['0.005'])):
    ax2.plot(x,[rt_down[e][i] for e in rt_up.keys()],label=labellist2[i])
ax2.set_xlabel(r'$\beta~[yr^{-1}]$')
#ax2.title.set_text('Response times for shrinking glacier')
#ax2.set_title('Response times for shrinking glacier')
#ax2.set_ylabel(r'$\tau~[yr]$')
ax2.legend()
plt.savefig('RT.pdf')
```
